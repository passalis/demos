```python
image_provider = VideoReader(0)
for counter, img in enumerate(image_provider):
    boxes = learner.infer(img, threshold=0.6)

    
    if len(boxes.data) > 0:
        bbox = BoundingBoxListToNumpyArray()(boxes)[0][:4].astype(np.int)
        face_img = img[bbox[1]:bbox[3],bbox[0]:bbox[2], :]

        results = recognizer.infer(face_img)

        cv2.rectangle(img, (bbox[0], bbox[1]), (bbox[2], bbox[3]), (0, 255, 0), 2)
        if results.confidence is not None and results.confidence > 0.5:
            cv2.putText(img, results.description , (bbox[0], max(bbox[1]-10, 0)),  cv2.FONT_HERSHEY_SIMPLEX, 0.8, (0, 255, 0), 1)
        else:
               cv2.putText(img, "Not found", (bbox[0], max(bbox[1]-10, 0)),  cv2.FONT_HERSHEY_SIMPLEX, 0.8, (255, 0, 0), 1)


    cv2.imshow('Result', img)
    if cv2.waitKey(1) == ord('q'):
        break
    
cv2.destroyAllWindows()
image_provider.close()

Jupyter Notebook
3. Face recognition Last Checkpoint: a few seconds ago (unsaved changes) Current Kernel Logo 

Python 3 (ipykernel)

    File
    Edit
    View
    Insert
    Cell
    Kernel
    Widgets
    Help

import cv2

%matplotlib inline

from matplotlib import pyplot as plt

import numpy as np





First, we need to load our model. In this tutorial we are using the Face Recognizer from OpenDR:

from opendr.perception.face_recognition.face_recognition_learner import FaceRecognitionLearner



We need to initialize our recognizer:

recognizer = FaceRecognitionLearner(device='cuda', backbone='mobilefacenet', mode='backbone_only')

Note that we can alter the device (e.g., 'cpu', 'cuda', etc.), on which the model runs, the backbone architecture that we are using, as well as a number of additional parameters that can make our model either faster or more accurate.

After creating our model, we need to download and load the pre-trained weights:

recognizer.download(path=".")

recognizer.load(".")

Model already exists
Loading backbone 'mobilefacenet'

We need a database of a few persons to recognize:

%ls persons

 me/         'Person 2'/  'Person 4'/  'Person 6'/
'Person 1'/  'Person 3'/  'Person 5'/

%ls persons/Person\ 4

1.jpg  2.jpg

Let's check a person's image:

from os.path import join

image_path = join("persons", "Person 4", "2.jpg")

img = cv2.imread(image_path)

plt.imshow(img[:, :, [2, 1, 0]])

<matplotlib.image.AxesImage at 0x7fca25c65c10>

Check our saved image

image_path = join("persons", "me", "1.jpg")

img = cv2.imread(image_path)

plt.imshow(img[:, :, [2, 1, 0]])

<matplotlib.image.AxesImage at 0x7fca25bca910>

Grab a frame from the camera and add it to the database:

from os.path import join

recognizer.fit_reference(path=join("persons"), save_path=".")







from opendr.perception.object_detection_2d.retinaface.retinaface_learner import RetinaFaceLearner

from opendr.perception.object_detection_2d.datasets.transforms import BoundingBoxListToNumpyArray

from utils import VideoReader

learner = RetinaFaceLearner(backbone='resnet', device='cuda')

learner.download(".", mode="pretrained")

learner.load("./retinaface_resnet")



Downloading pretrained files for retinaface_resnet
Downloading pretrained model...
Downloading metadata...
Downloading params...
Downloading symbol...
Loading model from path:  ./retinaface_resnet/retinaface_resnet

[14:55:21] ../src/base.cc:80: cuDNN lib mismatch: linked-against version 8101 != compiled-against version 8100.  Set MXNET_CUDNN_LIB_CHECKING=0 to quiet this warning.

Loaded mxnet model.

from opendr.perception.object_detection_2d.utils.vis_utils import draw_bounding_boxes



import numpy as np

image_provider = VideoReader(0)

for counter, img in enumerate(image_provider):

    boxes = learner.infer(img, threshold=0.6)



    

    if len(boxes.data) > 0:

        bbox = BoundingBoxListToNumpyArray()(boxes)[0][:4].astype(np.int)

        face_img = img[bbox[1]:bbox[3],bbox[0]:bbox[2], :]



        results = recognizer.infer(face_img)



        

        if results.confidence is not None and results.confidence > 0.5:
            cv2.putText(img, results.description , (bbox[0], max(bbox[1]-10, 0)),  cv2.FONT_HERSHEY_SIMPLEX, 0.8, (0, 255, 0), 1)
            cv2.rectangle(img, (bbox[0], bbox[1]), (bbox[2], bbox[3]), (0, 255, 0), 2)            
        else:
            cv2.putText(img, "Not found", (bbox[0], max(bbox[1]-10, 0)),  cv2.FONT_HERSHEY_SIMPLEX, 0.8, (0, 0, 255), 1)
            cv2.rectangle(img, (bbox[0], bbox[1]), (bbox[2], bbox[3]), (0, 0, 255), 2)





    cv2.imshow('Result', img)

    if cv2.waitKey(1) == ord('q'):

        break

    

cv2.destroyAllWindows()

image_provider.close()
```
